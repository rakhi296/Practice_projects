# Visualization of Corruption in Regions 

```python
# corruption in regions
corruption = df.groupby('Region')[['Trust (Government Corruption)']].mean()
corruption
```

```python
plt.rcParams['figure.figsize'] = (10,8)
region = df['Region']
corruption_ = df['Trust (Government Corruption)']
sns.barplot(x=region, y=corruption_)
plt.title('Perception of Corruption in various Region', fontsize='12')
plt.xlabel('Region', fontsize='12')
plt.ylabel('Corruption Index', fontsize='12')
plt.xticks(rotation=30, ha='right')
plt.show()
```

It shows that Austrailia and New Zealand has the highest perception of corruption, and Central and Eastern Europe has the least perception of corruption.

# Top 10 Happiest Countries Life Expectancy and 10 Least Happy Countries Life Expectancy

```python
# top 10 happiest countries based on life expectancy
top_10=df.sort_values(by='Health (Life Expectancy)', ascending=False).head(10)
top_10
```

```python
plt.rcParams['figure.figsize'] = (10,6)
hap_s = df['Happiness Score']
sns.scatterplot(x=df.Freedom, y=hap_s, hue=df.Region, s=100)
plt.legend(loc='upper left', fontsize='8')
plt.xlabel('Freedom')
plt.ylabel('Happiness Score')
plt.show()
```

```python
country = df.sort_values(by='Trust (Government Corruption)').head(10)
plt.rcParams['figure.figsize'] = (10,5)
plt.title('Countries with Least Corruption')
plt.xlabel('Country', fontsize=13)
plt.ylabel('Corruption Index', fontsize=13)
plt.xticks(rotation=30, ha='right')
corr_ = country['Trust (Government Corruption)']
coun_ =country['Country']
sns.barplot(x=coun_, y=corr_)
```

It shows that Indonesia has the lowest corruption

```python
country = df.sort_values(by='Trust (Government Corruption)').tail(10)
plt.rcParams['figure.figsize'] = (10,5)
plt.title('Countries with Most Corruption')
plt.xlabel('Country', fontsize=13)
plt.ylabel('Corruption Index', fontsize=13)
plt.xticks(rotation=30, ha='right')
corr_ = country['Trust (Government Corruption)']
coun_ =country['Country']
sns.barplot(x=coun_, y=corr_)
```

It shows that Rwanda has the highest corruption.

```python
# corruption vs happiness
plt.rcParams['figure.figsize'] = (10,6)
hap_s = df['Happiness Score']
cor_ = df['Trust (Government Corruption)']
sns.scatterplot(x=hap_s, y=cor_, hue=df.Region, s=100)
plt.legend(loc='upper left', fontsize='8')
plt.xlabel('Happiness Score')
plt.ylabel('Corruption')
plt.show()
```
